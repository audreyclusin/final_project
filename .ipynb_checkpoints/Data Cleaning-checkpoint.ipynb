```python
#Positive 60 Degrees Cleaning
```

```python
import pandas as pd
import numpy as np
```

```python
#Importing data for +60 degrees Latitude as a Pandas Dataframe
s = pd.read_csv('NOAA_Solar_Calculations - 60 Latitude.csv')
```

```python
s.shape
```

```python
print(s.columns)
```

```python
s.head()
```

```python
#Dropping unnecessary data 
s = s.drop(columns = ['NOAA Solar Calculations - Change any of the highlighted cells to get solar position data for that location and time-of-day for a year.', 'Unnamed: 1', 'Unnamed: 2', 'Time (hrs past local midnight)', 'Julian Day', 'Julian Century', 'Unnamed: 7', 'Geom Mean Long Sun (deg)', 'Geom Mean Anom Sun (deg)', 'Sunset Time (LST)', 'Sunlight Duration (minutes)', 'True Solar Time (min)', 'Hour Angle (deg)', 'Solar Elevation Angle (deg)', 'Approx Atmospheric Refraction (deg)', 'Unnamed: 34', 'Eccent Earth Orbit', 'Sun Eq of Ctr', 'Sun True Long (deg)', 'Sun True Anom (deg)', 'Sun Rad Vector (AUs)', 'Sun App Long (deg)', 'Mean Obliq Ecliptic (deg)', 'Obliq Corr (deg)', 'Sun Rt Ascen (deg)', 'Sun Declin (deg)', 'var y', 'Eq of Time (minutes)', 'HA Sunrise (deg)', 'Solar Noon (LST)', 'Sunrise Time (LST)'])
```

```python
s.head()
```

```python
#Extracting only one data point per week 
s = s.iloc[::7]
s = s.iloc[0:52]
```

```python
#Checking that the dataset is as long as it should be
print(len(s))
```

```python
#Saving cleaned data to a new file 
s.to_csv('Sixty_Degrees_Latitude.csv', index=False)
```
